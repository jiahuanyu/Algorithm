```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
random.seed(1)
```

```python
m = 50
sampleSizes = []
samplePrices = []
for i in range(m):
    sampleSizes.append(random.random() * 100 )
    samplePrices.append(random.random() * 3 * sampleSizes[i])
```

```python
plt.figure()
plt.scatter(sampleSizes,samplePrices)
```

```python
theta0 = 0
theta1 = 0
theta2 = 0
alpha = 0.00000001

for i in range(100000):
    sum_theta0 = 0
    sum_theta1 = 0
    sum_theta2 = 0
    for j in range(m):
        sum_theta0 = sum_theta0 + alpha * (theta0 + theta1 * sampleSizes[j] + theta2 * sampleSizes[j] * sampleSizes[j] - samplePrices[j]) * 1
        sum_theta1 = sum_theta1 + alpha * (theta0 + theta1 * sampleSizes[j] + theta2 * sampleSizes[j] * sampleSizes[j] - samplePrices[j]) * sampleSizes[j]        
        sum_theta2 = sum_theta2 + alpha * (theta0 + theta1 * sampleSizes[j] + theta2 * sampleSizes[j] * sampleSizes[j] - samplePrices[j]) * sampleSizes[j] * sampleSizes[j]
        
    theta0 = theta0 - sum_theta0 / m
    theta1 = theta1 - sum_theta1 / m
    theta2 = theta2 - sum_theta2 / m
    
print theta0,theta1,theta2
```

```python
plt.figure()
x = np.linspace(0,120,50)
y = theta0 + theta1 * x + theta2 * x * x
plt.plot(x,y,color='red')
plt.scatter(sampleSizes,samplePrices)
```
