```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 创建-3到3的50个点
x = np.linspace(-3,3,50)
# 线性方程
y1 = 2*x + 1
y2 = x**2
```

```python
plt.figure()
plt.plot(x,y1)
```

```python
plt.figure()
plt.plot(x,y1,color='red',linestyle='--',linewidth=2.0,label='up')
plt.plot(x,y2,label='down')
# 设置X轴范围
plt.xlim((-1,2))
# 设置Y轴范围
plt.ylim((-2,3))
# 设置x轴的标识
plt.xlabel('x-axis')
# 设置y轴的标识
plt.ylabel('y-axis')
# x轴坐标单元
x_ticks = np.linspace(-1,2,5)
plt.xticks(x_ticks)
# y轴坐标单元
plt.yticks([-2, -1, 1.22, 3],[r'$really\ bad$', r'$\alpha \ normal$', 'good', 'really good'])
# gca = get current axis
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

plt.legend(loc='lower right')
```
